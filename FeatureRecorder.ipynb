```python
import os
import pyaudio
import wave
import numpy as np
import IPython.display as ipd
```

```python
f = open('counter.txt', 'r')
counter = int(f.read())
f.close()
counter += 1

CHUNK = 1024
FORMAT = pyaudio.paInt16
CHANNELS = 1
RATE = 44100
RECORD_SECONDS = 1.5
WAVE_OUTPUT_FILENAME = r"orig/Без шума/Старт/{}.wav".format(counter)

p = pyaudio.PyAudio()

stream = p.open(format=FORMAT,
                channels=CHANNELS,
                rate=RATE,
                input=True,
                frames_per_buffer=CHUNK)

print("* recording ")

frames = []

for i in range(0, int(RATE / CHUNK * RECORD_SECONDS)):
    data = stream.read(CHUNK)
    frames.append(data)

print("* done recording ")

stream.stop_stream()
stream.close()
p.terminate()

wf = wave.open(WAVE_OUTPUT_FILENAME, 'wb')
wf.setnchannels(CHANNELS)
wf.setsampwidth(p.get_sample_size(FORMAT))
wf.setframerate(RATE)
wf.writeframes(b''.join(frames))
wf.close()


f = open('counter.txt', 'w')
f.write(str(counter))
f.close()

ipd.Audio(WAVE_OUTPUT_FILENAME, rate=RATE)
```
